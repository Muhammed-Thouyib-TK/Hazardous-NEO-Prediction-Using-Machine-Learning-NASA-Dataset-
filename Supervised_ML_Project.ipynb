```python
# to create a function that remove outliers

def remove_outliers(data,column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - IQR*1.5
    upper = Q3 + IQR*1.5
    before = len(data)
    data = data[(data[column]>=lower) & (data[column]<=upper)]
    Updated = before - len(data)
    print('Removed rows :',Updated)
```

```python
for i in df.drop('is_hazardous',axis=1).columns:
    remove_outliers(df,i)
```

```python
# to split the dataset

x = df.drop('is_hazardous',axis=1)
y = df['is_hazardous']
```

```python
# to split the data into training & testing (CV)

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# to check the training data is balanced

y_train.value_counts().plot(kind='bar')
```

```python
# to generate some fake data based on the input data we have

from imblearn import over_sampling

# synthetic minority over-sampling technique (SMOTE)

se = over_sampling.SMOTE()
```

```python
x_train_balanced,y_train_balanced = se.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
y_train_balanced.value_counts()
```

```python
x_train_balanced
```

```python
x_test
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(x_train_balanced,y_train_balanced)
```

```python
y_pred_test = lr.predict(x_test)
y_pred_test
```

```python
# to get the classification report

from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_test))
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train_balanced,y_train_balanced)
```

```python
rfc_pred_test = rf.predict(x_test)
```

```python
print(classification_report(y_test,rfc_pred_test))
```

```python
from sklearn.metrics import roc_auc_score

print("ROC-AUC Score:", roc_auc_score(y_test, rf.predict_proba(x_test)[:, 1]))
```
